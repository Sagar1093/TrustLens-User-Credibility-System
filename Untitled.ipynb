```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import MinMaxScaler
import pickle
```

```python
main_df = pd.read_csv("Ecommerce_Consumer_Behavior_Analysis_Data.csv")
main_df.head(10)
```

```python
main_df.columns
```

```python
unwanted_columns=["Age","Gender","Marital_Status","Education_Level","Occupation","Purchase_Category",
                  "Location","Device_Used_for_Shopping","Purchase_Channel","Payment_Method","Time_of_Purchase",
                  "Social_Media_Influence","Engagement_with_Ads","Income_Level","Discount_Sensitivity","Brand_Loyalty",
                  "Discount_Used","Purchase_Intent","Shipping_Preference","Customer_Loyalty_Program_Member"]
```

```python
df = main_df.drop(unwanted_columns,axis =1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape[0]
```

```python
df.dtypes
```

```python
df["Purchase_Amount"] = df["Purchase_Amount"].str.replace("$","")
df["Purchase_Amount"] = df["Purchase_Amount"].astype(float)

```

```python
df.dtypes
```

```python
plt.scatter(
    df['Frequency_of_Purchase'], 
    df['Product_Rating'],
   
)
```

```python

for _ in range(20):
    sample = {
        "Purchase_Amount": np.random.normal(300, 20),
        "Frequency_of_Purchase": np.random.normal(5, 1),
        "Product_Rating": np.random.choice([1,2]),
        "Return_Rate": np.random.normal(4, 1),
        "Customer_Satisfaction": np.random.choice([0,1])
    }
    df = pd.concat([df, pd.DataFrame([sample])], ignore_index=True)

```

```python
df.fillna({
    "Customer_Satisfaction": df["Customer_Satisfaction"].mean(),
    "Product_Rating": df["Product_Rating"].median(),
    "Frequency_of_Purchase": 1
}, inplace=True)
```

```python
df.head()
```

```python
cols = ["Frequency_of_Purchase","Product_Rating","Time_Spent_on_Product_Research(hours)","Return_Rate","Customer_Satisfaction"]
for col in cols:
    df[col] = df[col] + np.random.normal(0,0.5,len(df))




```

```python
df.loc[df["Purchase_Amount"] > 1000 , "Purchase_Amount"] /= 10
```

```python
df["value_per_visit"] = df["Purchase_Amount"]/(df["Frequency_of_Purchase"]+1)
df["return_ratio"] = df["Return_Rate"]/(df["Frequency_of_Purchase"]+2)
```

```python
for col in ["return_ratio", "value_per_visit"]:
    df[col] += np.random.normal(0, 0.02, len(df))

df["return_ratio"] += np.random.normal(0, 0.01, len(df))
df["return_ratio"] = df["return_ratio"].clip(0, 1)
```

```python
df.loc[df["Return_Rate"]>2,"Customer_Satisfaction"]-= 2
df.loc[df["Time_Spent_on_Product_Research(hours)"]>5,"Product_Rating"] +=1
```

```python
outliers = df.sample(30,random_state = 42)

df.loc[outliers.index,"Return_Rate"] *=3
df.loc[outliers.index,"Frequency_of_Purchase"] *= 0.2
df["Return_Rate"] = df["Return_Rate"].replace(0, np.random.uniform(0.1, 0.5))
df["Return_Rate"] = df["Return_Rate"].clip(0)
df["Product_Rating"] = df["Product_Rating"].clip(1,5)
df["Customer_Satisfaction"] = df["Customer_Satisfaction"].clip(0,10)
df["return_ratio"] = df["return_ratio"].clip(0,1.5)
```

```python
plt.scatter(
    df['Frequency_of_Purchase'], 
    df['Product_Rating'],
   
)
```

```python

df["Product_Rating"] = df["Product_Rating"].round().astype(int)
```

```python
df.dtypes
```

```python
features = df[["Product_Rating","Frequency_of_Purchase",
               "Customer_Satisfaction","value_per_visit","return_ratio"]]
```

```python
scaler = StandardScaler()
```

```python
scaled_features = scaler.fit_transform(features)
```

```python
model = IsolationForest(contamination=0.042,n_estimators=250, max_samples='auto')
```

```python
model.fit(scaled_features)
```

```python
df["anomaly"] = model.predict(scaled_features)
df.head()
```

```python
plt.scatter(
    df['Customer_Satisfaction'], 
    df['return_ratio'],
    c=df["anomaly"]
)

df['anomaly'].value_counts()
```

```python
scores = model.decision_function(scaled_features)
df['anomaly_score'] = scores
```

```python
df[["anomaly","anomaly_score"]].sort_values(by = "anomaly_score")
```

```python
df.head(2)
```

```python
plt.hist(df['anomaly_score'], bins=100)
plt.title("Anomaly Score Distribution")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

```python
scoring_features = ["Product_Rating","Frequency_of_Purchase",
               "Customer_Satisfaction","value_per_visit","return_ratio"]
```

```python
score_scaler = MinMaxScaler()
df[scoring_features] = score_scaler.fit_transform(df[scoring_features])
```

```python
df["Trust_Score"] = (
    0.35*df["Customer_Satisfaction"]+
    0.25*df["Product_Rating"]+
    0.20*df["Frequency_of_Purchase"]-
    0.35*(1-df["return_ratio"])-
    0.3*(1-df["value_per_visit"])

)
```

```python
df.loc[df["anomaly"] == -1,"Trust_Score"] /= 6
df["Trust_Score"] = df["Trust_Score"] + df["anomaly_score"]
```

```python
df["Trust_Score"] = (df["Trust_Score"] - df["Trust_Score"].min()) / (df["Trust_Score"].max() - df["Trust_Score"].min())
df["Trust_Score"] = df["Trust_Score"]*100
```

```python
def label(score):
    if score > 60:
        return "High Trust"
    elif score > 35:
        return "Medium Trust"
    else:
        return "Low Trust"

df["Trust_Label"] = df["Trust_Score"].apply(label)
```
